## Krig data to 70x50 rectangular grid

```python
lon_linspace = np.linspace(min_lon, max_lon, 70)
lat_linspace = np.linspace(min_lat, max_lat, 50)

new_tmax = np.empty((np.shape(tmax_sub)[0], np.shape(lon_linspace)[0], np.shape(lat_linspace)[0]), float)
print(np.shape(lon_linspace), np.shape(lat_linspace), np.shape(new_tmax))
```

```python
''' Create tree '''
print(np.array([lon_sub, lat_sub]).T.reshape(-1,2))

tree = KDTree(np.array([lon_sub, lat_sub]).T.reshape(-1,2))
```

```python
def DoKrig(x, y, val, x_target, y_target):
    OK = OrdinaryKriging(
                x, 
                y, 
                val, 
                variogram_model='spherical',
                verbose=False,
                enable_plotting=False,
                nlags=10,
                )
    return OK.execute("points", x_target, y_target)
```

```python
num_neighbors = 2
new_lon = np.empty((np.shape(lon_linspace)[0],
                    np.shape(lat_linspace)[0]), float)
new_lat = np.empty((np.shape(lon_linspace)[0],
                    np.shape(lat_linspace)[0]), float)
```

```python
for i in range(np.shape(new_tmax)[1]):
    for j in range(np.shape(new_tmax)[2]):
        print("-----------------", i, j, "------------------")
        
        ''' Start populating new_lon, new_lat '''
        new_lon[i][j] = lon_linspace[i]
        new_lat[i][j] = lat_linspace[j]
        
        ''' Get closest neighbor(s) to desired point in tmax '''
        target_point = np.array([lon_linspace[i], lat_linspace[j]])
    
        dis, idx = tree.query(target_point, k=num_neighbors)
    
        lon_krig = lon_sub[idx]
        lat_krig = lat_sub[idx]
    
        for t in range(np.shape(new_tmax)[0]):
            tmax_krig = tmax_sub[t][idx]
        
            if not (np.all(tmax_krig == tmax_krig[0])): # Array elements are different

                zstar, ss = DoKrig(lon_krig, lat_krig, tmax_krig, new_lon[i], new_lat[j])

                new_tmax[t][i][j] = zstar[0]
                #print("tmax_krig =", tmax_krig, ", new_tmax[t][i][j] =", round(new_tmax[t][i][j], 3),
                #     ", lon_krig =", lon_krig, ", target =", new_lon[i][j],
                #     ", lat_krig =", lat_krig, ", target =", new_lat[i][j])

            else: # All array elements are the same
                #print(tmax_krig[idx[0]])
                if not (-999 <= tmax_krig[0] and tmax_krig[0] <= 367):
                    #print("AHHHHHHH | tmax_krig =", tmax_krig, ", new_tmax[t][i] =", new_tmax[t][i])
                    new_tmax[t][i][j] = 0
                else:
                    new_tmax[t][i][j] = tmax_krig[0]
                    #print("tmax_krig =", tmax_krig, ", new_tmax[t][i] =", new_tmax[t][i])


            #print("z =", new_tmax[t][i], ", tmax =", tmax_krig, "i =", i, ", t =", t, ", idx =", idx, ", point =", point, " lon =", lon_krig, ", lat =", lat_krig)
```
